```python
def amostragem_aleatoria(df:pd.DataFrame,amostras:int,reposicao:bool = True):
    linhas,colunas = df.shape
    colunas = df.keys()
    df_amostras = pd.DataFrame(columns=colunas)
    if amostras == 0 or df.empty:
        return df_amostras
    if amostras > len(df):
        raise Exception("Nro Amostras maior que a população!")
    if reposicao:
        for i in range(amostras):
            ind = randint(0,linhas)
            amostra = df.iloc[ind]
            df_amostras.loc[len(df_amostras)] = amostra
    else: #sem reposicao
        indices_amostrados = []
        for i in range(amostras):
            linhas,cols = df.shape
            ind = randint(0,linhas-1)
            while ind in indices_amostrados:
                ind = randint(0,linhas-1)
            indices_amostrados.append(ind)
            amostra = df.iloc[ind]
            df_amostras.loc[len(df_amostras)] = amostra
    return df_amostras

def amostragem_estratificada(df:pd.DataFrame,amostras:int,coluna:str):
    dicionario = {}
    dicionario_amostra = {}
    conteudos = df[coluna].unique()
    linhas,cols = df.shape
    porcentagem = float(amostras)/float(linhas)
    for conteudo in conteudos:
        qtd = df[coluna].value_counts()[conteudo]
        dicionario[conteudo] = qtd
        qtd_amostra = qtd * porcentagem
        dicionario_amostra[conteudo] = round(qtd_amostra)
    # em caso de erro de arredondamento, a chave com maior quantidade de amostras sera afetada, sempre!
    while sum(dicionario_amostra.values()) > amostras or sum(dicionario_amostra.values()) < amostras:
        print("entrou")
        maior_amostra = max(dicionario_amostra.values())
        for chave,valor in dicionario_amostra.items():
            if valor == maior_amostra:
                match(sum(dicionario_amostra.values())>amostras):
                    case True:dicionario_amostra[chave] = valor - 1
                    case False: dicionario_amostra[chave] = valor + 1
    
    
            
    
```
